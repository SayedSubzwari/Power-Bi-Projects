```python
import pandas as pd
import plotly
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
#reading file
df = pd.read_csv('/Users/vaibhaviawghate/Documents/Data Projects/countries of the world.csv')
df.head()
```

```python
# replacing NaN with 0 and ',' with '.'
df.fillna(0, inplace=True)
for column in df.columns:
    try:
        df[column] = df[column].str.replace(",", ".").astype(float)
    except:
        print('')
```

```python
df.describe()
```

```python
#plot to show population by country
data = dict(type='choropleth',
locations = df['Country'],
locationmode = 'country names', z = df['Population'],
text = df['Country'], colorbar = {'title':'Population'},
colorscale = [[0,"rgb(5, 10, 172)"],[0.35,"rgb(40, 60, 190)"],[0.5,"rgb(70, 100, 245)"],\
            [0.6,"rgb(90, 120, 245)"],[0.7,"rgb(106, 137, 247)"],[1,"rgb(220, 220, 220)"]],
reversescale = True) 
layout = dict(title='Population by Country',
geo = dict(showframe=False,projection={'type':'Mercator'}))
choromap = go.Figure(data = [data],layout = layout)
py.iplot(choromap,validate=False)
```

```python
# plot to show GDP by country 
data = dict(type='choropleth',
locations = df['Country'],
locationmode = 'country names', z = df['GDP ($ per capita)'],
text = df['Country'], colorbar = {'title':'GDP ($ per capita)'},
colorscale = [[0.0, 'rgb(242,240,247)'],[0.2, 'rgb(218,218,235)'],[0.4, 'rgb(188,189,220)'],\
            [0.6, 'rgb(158,154,200)'],[0.8, 'rgb(117,107,177)'],[1.0, 'rgb(84,39,143)']],
reversescale = False) 
layout = dict(title='Wealthy Countries',
geo = dict(showframe=False,projection={'type':'Mercator'}))
choromap = go.Figure(data = [data],layout = layout)
py.iplot(choromap,validate=False)
```
